### Executive Summary
Patients with Cystic Fibrosis (CF) face a native lifetime kidney stone prevalence of up to $32\%$ due to genetic and metabolic abnormalities, including enteric hyperoxaluria and hypocitraturia (Terribile et al., 2006). When these patients reside in the warm, humid geographic regions known as the "Kidney Stone Belt," seasonal ambient heat multipliers drastically accelerate fluid loss (Brikowski et al., 2008; Dallas et al., 2017).

This Python-based **Clinical Action Engine** bridges a critical healthcare gap. Instead of reacting to kidney stones after they form, this program dynamically evaluates environmental factors (temperature and humidity) alongside a patient's biochemical markers to calculate active risk levels. It then automatically generates highly tailored, evidence-based treatment directives for the care team in real time.

````python
def calculate_sweat_loss_f(temperature_f, humidity):
    if temperature_f <= 77:
        return 0.0
    else:
        # Humidity is a percentage (like 80%), so we divide by 100 to get a decimal (0.80)
        humidity_decimal = humidity / 100.0

        # Calculate the degrees above 77°F
        degrees_above_limit = temperature_f - 77

        # Multiply it all together: degrees over limit * 0.028 liters * humidity decimal
        total_loss = degrees_above_limit * 0.028 * humidity_decimal

        """
        --- BIOLOGICAL AND ENVIRONMENTAL DERIVATION EXPLAINER ---

        1. THE COMFORT THRESHOLD:
           77°F (25°C) represents the upper limit of the human thermal neutral zone.
           At or below this temperature, sweating is clinically negligible (Brikowski et al., 2008).

        2. THE FAHRENHEIT SCALE FACTOR (0.028 L):
           Clinical baseline active sweat loss is estimated at 0.05 Liters per hour per 1°C
           above the neutral zone (Brikowski et al., 2008). To convert this rate to Fahrenheit,
           we divide 0.05 L by 1.8 (since 1°C = 1.8°F):
           0.05 / 1.8 = 0.02777... Liters, which rounds to exactly 0.028 Liters per 1°F.

        3. THE HUMIDITY MULTIPLIER:
           High atmospheric moisture severely blunts the efficiency of evaporative cooling.
           The sweat glands must produce a higher volume of fluid in muggy weather to attempt
           the same degree of core cooling (Dallas et al., 2017).
        """

        # Round the answer to 2 decimal places so it looks clean
        return round(total_loss, 2)

# Let's test our formula with a hot Southern summer day: 95°F and 80% humidity
test_temp = 95
test_humidity = 80

water_lost = calculate_sweat_loss_f(test_temp, test_humidity)

print("On a hot day of", test_temp, "°F and", test_humidity, "% humidity...")
print("The patient loses", water_lost, "Liters of water just from sweating!")
````

````python
# First, we keep our previous formulas so this cell is fully self-contained!
def calculate_sweat_loss_f(temperature_f, humidity):
    if temperature_f <= 77:
        return 0.0
    else:
        humidity_decimal = humidity / 100.0
        degrees_above_limit = temperature_f - 77
        total_loss = degrees_above_limit * 0.028 * humidity_decimal
        return round(total_loss, 2)

def calculate_hydration_balance(starting_water, water_drunk, temperature_f, humidity):
    sweat_lost = calculate_sweat_loss_f(temperature_f, humidity)
    final_balance = starting_water + water_drunk - sweat_lost
    return round(final_balance, 2)

# --- OUR NEW CLINICAL EVALUATION ENGINE ---
def evaluate_stone_risk(final_water, citrate_mg, hyperoxaluria_severity):
    """
    --- BIOLOGICAL AND ENVIRONMENTAL DERIVATION EXPLAINER ---

    1. DEHYDRATION CRITERIA (< 2.0 L):
       Urinary volume is the primary physical defense against crystallization.
       A final water balance under 2.0 Liters signals highly concentrated urine,
       drastically lowering the threshold needed for salts to precipitate (Dallas et al., 2017).

    2. INHIBITOR DEFICIT (< 320 mg/day):
       Citrate is the body's premier natural crystallization inhibitor. It binds to calcium,
       preventing it from pairing with oxalate. Levels below 320 mg/day (hypocitraturia)
       leave the renal tubules chemically defenseless (Shastri et al., 2023).

    3. ENTERIC HYPEROXALURIA ("Severe"):
       Malabsorption in Cystic Fibrosis leaves free fats in the gut, which bind to calcium.
       This leaves dietary oxalate unbound and hyper-absorbed across the colon.
       A "Severe" grade represents an overwhelming metabolic mineral load in the kidneys (Knauf et al., 2016).
    """

    # Track how many clinical red flags are triggered
    red_flags = 0

    if final_water < 2.0:
        red_flags += 1

    if citrate_mg < 320:
        red_flags += 1

    if hyperoxaluria_severity == "Severe":
        red_flags += 1

    # Assign risk level based on the number of red flags triggered
    if red_flags == 3:
        return "CRITICAL RISK (Perfect Storm: Dehydrated, Low Inhibitors, High Oxalate!)"
    elif red_flags == 2:
        return "HIGH RISK (Two clinical defense mechanisms have failed.)"
    elif red_flags == 1:
        return "MODERATE RISK (One clinical defense mechanism has failed.)"
    else:
        return "LOW RISK (Patient parameters are currently stable.)"

# Let's test a real, highly vulnerable Cystic Fibrosis patient:
# They start the day with 2.5 Liters of water, and drink 1.2 Liters.
# They are out in the Kidney Stone Belt on a hot, muggy day (95°F and 80% humidity).
# Their lab work shows a citrate deficit (280 mg) and severe enteric hyperoxaluria.

start_water = 2.5
drunk_water = 1.2
current_temp = 95
current_humidity = 80
patient_citrate = 280
patient_oxalate = "Severe"

# 1. Calculate how much water they have left
ending_water = calculate_hydration_balance(start_water, drunk_water, current_temp, current_humidity)

# 2. Run our risk evaluator
risk_assessment = evaluate_stone_risk(ending_water, patient_citrate, patient_oxalate)

print("--- CLINICAL RISK SCREENING ---")
print("Ending Water Level in Body:", ending_water, "L")
print("Patient Urinary Citrate Level:", patient_citrate, "mg/day")
print("Enteric Hyperoxaluria Severity:", patient_oxalate)
print("--------------------------------")
print("EVALUATION RESULT:", risk_assessment)
````

````python


# --- 1. PREVIOUS CLIMATE & HYDRATION FORMULAS ---
def calculate_sweat_loss_f(temperature_f, humidity):
    if temperature_f <= 77:
        return 0.0
    else:
        humidity_decimal = humidity / 100.0
        degrees_above_limit = temperature_f - 77
        total_loss = degrees_above_limit * 0.028 * humidity_decimal
        return round(total_loss, 2)

def calculate_hydration_balance(starting_water, water_drunk, temperature_f, humidity):
    sweat_lost = calculate_sweat_loss_f(temperature_f, humidity)
    final_balance = starting_water + water_drunk - sweat_lost
    return round(final_balance, 2)

# --- 2. RISK EVALUATION ENGINE ---
def evaluate_stone_risk(final_water, citrate_mg, hyperoxaluria_severity):
    red_flags = 0
    if final_water < 2.0:
        red_flags += 1
    if citrate_mg < 320:
        red_flags += 1
    if hyperoxaluria_severity == "Severe":
        red_flags += 1

    if red_flags == 3:
        return "CRITICAL RISK"
    elif red_flags == 2:
        return "HIGH RISK"
    elif red_flags == 1:
        return "MODERATE RISK"
    else:
        return "LOW RISK"

# --- 3. OUR NEW CLINICAL ACTION ENGINE ---
def generate_clinical_recommendations(risk_level, ending_water, hyperoxaluria_grade):
    """
    --- BIOLOGICAL AND ENVIRONMENTAL DERIVATION EXPLAINER ---

    1. THE >3.0L FLUID STANDARD:
       In hot climate zones, sweat drains body water rapidly. Standard 2.0L targets fail (Brikowski et al., 2008).
       An escalated target of >3.0L of daily fluid intake is clinically mandated to force
       dilution in vulnerable, supersaturated kidneys (Prevalence and Risk Factors, 2026).

    2. THE MICRO MICROSCOPY SCREEN:
       Crystals naturally precipitate and grow as urine cools. Specimen analysis must occur
       within 2 hours of urination to avoid false crystalluria readings (StatPearls, 2024).

    3. THE PANCREATIC ENZYME AUDIT:
       Overmedicating exocrine pancreatic insufficiency with high enzyme doses introduces a massive
       purine overload, triggering dangerous iatrogenic uric acid spikes (Stapleton et al., 1976).
    """

    print("\n==================================================")
    print("📋 CLINICAL PRACTICE RECOMMENDATIONS")
    print("==================================================")

    if risk_level in ["CRITICAL RISK", "HIGH RISK"]:
        print("🚨 ACTION REQUIRED: ACTIVATE CLIMATE-SMART SURVEILLANCE PROTOCOL")

        # Recommendation 1: Hydration Goal Escalation
        print(f"\n1. FLUID PRESCRIPTION (Target: >3.0 L/day)")
        print(f"   - Current calculated urine volume is heavily compromised at {ending_water} L.")
        print("   - Action: Mandate intensive fluid targets to achieve >3.0L verified daily urine output.")

        # Recommendation 2: Microscopic Screening
        print("\n2. PROACTIVE MICRO-URINALYSIS (The Liquid Biopsy)")
        print("   - Action: Order a fresh microscopic urinalysis to screen for crystal morphology/size.")
        print("   - Note: Specimen must be processed within 2 hours of voiding to prevent cooling artifacts.")

        # Recommendation 3: Medication Check
        print("\n3. MEDICATION AUDIT (Pancreatic Enzymes)")
        print(f"   - Patient has '{hyperoxaluria_grade}' Enteric Hyperoxaluria.")
        print("   - Action: Meticulously audit current pancreatic enzyme dosages to rule out purine overload.")

    elif risk_level == "MODERATE RISK":
        print("⚠️ ACTION RECOMMENDED: ADVANCED MONITORING")
        print("\n- Increase daily fluid intake by 500 mL to prevent dehydration spikes.")
        print("- Schedule routine 24-hour metabolic urine panels to track citrate trends.")

    else:
        print("✅ STANDARD MAINTENANCE PROTOCOL")
        print("\n- Maintain baseline hydration guidelines.")
        print("- Schedule routine annual metabolic profiles during peak summer months.")

    print("==================================================")

# --- 4. RUNNING THE COMPLETE SYSTEM ---
# Let's test our highly vulnerable patient from Step 4 again:
start_water = 2.5
drunk_water = 1.2
current_temp = 95
current_humidity = 80
patient_citrate = 280
patient_oxalate = "Severe"

# Step A: Run the hydration balance calculator
ending_water = calculate_hydration_balance(start_water, drunk_water, current_temp, current_humidity)

# Step B: Grade their medical stone risk
risk_level = evaluate_stone_risk(ending_water, patient_citrate, patient_oxalate)

# Step C: Print the diagnostic screening
print("--- DIAGNOSTIC PATIENT SCREENING ---")
print("Ending Water Level in Body:", ending_water, "L")
print("Patient Urinary Citrate Level:", patient_citrate, "mg/day")
print("Enteric Hyperoxaluria Severity:", patient_oxalate)
print("Calculated Risk Status:", risk_level)

# Step D: Run our new clinical action engine to print automated directives!
generate_clinical_recommendations(risk_level, ending_water, patient_oxalate)
````
